## Introduction

The goal of this project is to design and implement a Big Data pipeline
to analyze historical trends in video games and to predict the expected
user rating of upcoming games.

The system combines real-time data ingestion with batch analytics
and simple machine learning models.


## System Architecture

The architecture of the system is composed of the following components:

- **Python crawler** to extract data from the IGDB API
- **Apache Kafka** for streaming data ingestion
- **Apache Spark Structured Streaming** for data processing
- **Elasticsearch** for data storage
- **Kibana** for data visualization

This architecture supports both retrospective analysis and future predictions.


## Dataset Description

The dataset contains information about video games, including:

- Game name
- Franchise / series
- Release date
- User rating and rating count
- Popularity indicators (hype)
